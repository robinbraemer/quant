```python

import json

import pandas as pd
import vectorbt as vbt
```

```python
df = pd.read_csv("TradingView_Alerts_Log_2021-09-26.csv", parse_dates=True)
df
```

```python
df = df[df['Description'].str.startswith("{\n") ]
df = pd.DataFrame.from_records(df['Description'].apply(json.loads))
cols = ['open', 'low', 'high', 'close', 'volume']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
df = df[df.auth.isna()]
cols = ['time','tf', 'side', 'ticker'] + cols
df = df[cols]
```

```python
df['time'] = pd.to_datetime(df['time'])
df.set_index("time",inplace=True)
df.sort_values(by="time", ascending=True, inplace=True)
print(df.dtypes)
df
```

```python
by_ticker = df.groupby(['ticker'])
by_ticker['tf'].value_counts()
```

```python
tf = "1m"
ticker = 'XRPUSDTPERP'

signals = df[(df.ticker == ticker) & (df['tf'] == tf)]

entries = signals['side'] == 'Buy'
exits = signals['side'] == 'Sell'

vbt.settings.portfolio.init_cash = 1000
pf = vbt.Portfolio.from_signals(signals['close'],
                                direction='both',
                                entries=entries,
                                exits=exits,
                                # short_entries=short_entries,
                                # short_exits=short_exits,
                                # tp_stop=0.03,
                                # sl_trail=0.3
                                )
print(pf.init_cash + pf.total_profit())
pf.plot().show()
```
